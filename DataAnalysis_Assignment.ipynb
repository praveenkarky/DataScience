# 1.9  8) Test the similarity between the degree_t(Sci&Tech) and specialisa-tion(Mkt&HR)  with respect to salary at significance level of 5%.(Makedecision using Hypothesis Testing)

```python
from scipy.stats import ttest_ind
grp1= dataset[dataset["degree_t"] == 'Sci&Tech']["salary"]
grp2= dataset[dataset["specialisation"] =='Mkt&HR']["salary"]
res= ttest_ind(grp1,grp2)
res
```

Hypothesis Test (p<0.05):

There is no similarity between the degree_t(Sci&Tech) and specialisa-tion(Mkt&HR) respect to the salary.
The pvalue=0.007 is less then pvalue = 0.05 so NULL Hypothesis rejected.
